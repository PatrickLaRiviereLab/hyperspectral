```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from data_objects import get_fluorophore_list
```

```python
fluorophore_list = get_fluorophore_list()
#filter_bins = [
#    (425, 475), # 450/50
#    (522, 538), # 530/15
#    (600, 620), # 610/20
#    (630, 665)  # 630/69
#]

filter_bins = [
    (400, 470),
    (490, 560),
    (570, 630),
    (640, 900)
]
```

```python
color = ['tab:orange', 'tab:green', 'tab:red', 'tab:blue']
for i in range(len(fluorophore_list)):
    fluorophore = fluorophore_list[i]
    plt.plot(fluorophore.spectra.emission.index, fluorophore.spectra.emission, label=fluorophore.name, color=color[i])

color = ['tab:blue', 'tab:green', 'tab:orange', 'tab:red']
for i in range(len(filter_bins)):
    bin = filter_bins[i]
    height = bin[0]*0.020/900
    avg = (bin[0] + bin[1])//2
    bandwidth = (bin[1] - bin[0])
    label = "{}/{} bandwidth".format(avg, bandwidth)
    plt.plot(bin, (height, height), label=label, color=color[i])
    
plt.xlabel("Wavelength")
plt.ylabel("Probability Density")
plt.title("Emission Spectra with Filter Bandwidths")
plt.legend()

plt.savefig("Figures/Filter Bandwidths 3.png")

plt.show()
```
